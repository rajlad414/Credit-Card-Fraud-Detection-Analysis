```python
train_dataset.drop(train_dataset[train_dataset['Amount'] > 0.75].index, inplace = True)
test_dataset.drop(test_dataset[test_dataset['Amount'] > 0.75].index, inplace = True)
```

* We can Drop the Time column as we converted it the time interval of 48 hrs

```python
train_dataset.drop('Time', axis = 1, inplace = True)
test_dataset.drop('Time', axis = 1, inplace = True)
```

### **Handling Imbalance Data**

```python
X = train_dataset.drop('Class', axis = 1)
y = train_dataset['Class']
```

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state = 42)
X_resampled, y_resampled = sm.fit_resample(X, y)
```

```python
print(f"The shape of new resampled class is: {y_resampled.shape}")
```

```python
class_data = pd.Series(y_resampled).value_counts().to_dict()
print(f"The class count for each class is: {class_data}")
plt.bar(list(class_data.keys()), list(class_data.values()))
```

* After applying SMOTE method for oversampling the miniority data, we observe that the both the class are now balanced

### **Spliting the dataset**

* We split the dataset for the both imbalanced(for tree based model) and balanced data

```python
from sklearn.model_selection import train_test_split
x_resampled_train, x_resampled_test, y_resampled_train, y_resampled_test = train_test_split(X_resampled, y_resampled, test_size=0.3, random_state=42)
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

### **GausianNB Classifier**

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, precision_score, f1_score, recall_score, confusion_matrix

models = ['GaussianNB']
data = []
gnb = GaussianNB()
gnb.fit(x_resampled_train, y_resampled_train)
y_pred = gnb.predict(x_resampled_test)
data.append({
    'accuracy' : accuracy_score(y_resampled_test, y_pred).round(2),
    'precision' : precision_score(y_resampled_test, y_pred).round(2),
    'f1' : f1_score(y_resampled_test, y_pred).round(2),
    'recall' : recall_score(y_resampled_test, y_pred).round(2),
})
print(f"Confusion Matrix for GaussianNB:\n{confusion_matrix(y_resampled_test, y_pred)}")
model_data = pd.DataFrame(data, index = models)
model_data
```

### **LogisticRegression Classifier**

```python
from sklearn.linear_model import LogisticRegression
models.append('LogisticRegression')
lr = LogisticRegression()
lr.fit(x_resampled_train, y_resampled_train)
y_pred = lr.predict(x_resampled_test)
data.append({
    'accuracy' : accuracy_score(y_resampled_test, y_pred).round(2),
    'precision' : precision_score(y_resampled_test, y_pred).round(2),
    'f1' : f1_score(y_resampled_test, y_pred).round(2),
    'recall' : recall_score(y_resampled_test, y_pred).round(2),
})
print(f"Confusion Matrix for LogisticRegression:\n{confusion_matrix(y_resampled_test, y_pred)}")
model_data = pd.DataFrame(data, index = models)
model_data
```

### **SVC**

```python
from sklearn.svm import SVC
models.append('SVC')
svc = SVC()
svc.fit(x_train, y_train)
y_pred = svc.predict(x_test)
data.append({
    'accuracy' : accuracy_score(y_test, y_pred).round(2),
    'precision' : precision_score(y_test, y_pred).round(2),
    'f1' : f1_score(y_test, y_pred).round(2),
    'recall' : recall_score(y_test, y_pred).round(2),
})
print(f"Confusion Matrix for SVC:\n{confusion_matrix(y_test, y_pred)}")
model_data = pd.DataFrame(data, index = models)
model_data
```

### **Anomaly Detection**

```python
# Assuming the data follows a multivariate Gaussian distribution
from scipy.stats import multivariate_normal

# Fit the model on non-fraudulent transactions
normal_data = train_dataset[train_dataset['Class'] == 0][['time_in_hrs', 'Amount']]
mu, sigma = normal_data.mean(), normal_data.cov()

# Score each transaction
train_dataset['AnomalyScore'] = multivariate_normal.pdf(train_dataset[['time_in_hrs', 'Amount']], mean=mu, cov=sigma)

# Visualize scores
plt.figure(figsize=(10, 6))
plt.scatter(train_dataset.index, train_dataset['AnomalyScore'], c=train_dataset['Class'], cmap='viridis')
plt.title('Anomaly Scores for Fraudulent and Non-Fraudulent Transactions')
plt.xlabel('Transaction Index')
plt.ylabel('Anomaly Score')
plt.show()

# Find the threshold for marking a transaction as fraudulent
threshold = train_dataset[train_dataset['Class'] == 1]['AnomalyScore'].min()

# Use the anomaly score as a feature
x_train['AnomalyScore'] = multivariate_normal.pdf(x_train[['time_in_hrs', 'Amount']], mean=mu, cov=sigma)
x_test['AnomalyScore'] = multivariate_normal.pdf(x_test[['time_in_hrs', 'Amount']], mean=mu, cov=sigma)
```
